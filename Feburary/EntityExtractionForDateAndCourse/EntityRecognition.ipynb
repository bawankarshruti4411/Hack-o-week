```python
import re

def extract_entities(query):
    entities = {
        "semester": None,
        "course": None,
        "date": None
    }

    # Semester extraction
    sem_match = re.search(r'(?:sem|semester)\s*(\d+)', query, re.IGNORECASE)
    if sem_match:
        entities["semester"] = sem_match.group(1)

    # Course / department extraction
    course_match = re.search(r'\b(CS|IT|ME|CE|EE|EC)\b', query, re.IGNORECASE)
    if course_match:
        entities["course"] = course_match.group(1)

    # Date extraction
    date_match = re.search(r'\b(today|tomorrow|\d{1,2}\s*(jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec))', query, re.IGNORECASE)
    if date_match:
        entities["date"] = date_match.group(0)

    return entities


# Example Query
query = "When is SEM 5 CS exam?"

entities = extract_entities(query)
print(entities)
```

```python
def generate_response(entities):

    semester = entities["semester"]
    course = entities["course"]

    if semester and course:
        return f"The exam schedule for {course} Semester {semester} will be announced soon."

    elif semester:
        return f"Semester {semester} exam schedule will be released soon."

    else:
        return "Please specify semester or course for exam information."


response = generate_response(entities)
print(response)
```

```python
import re

# Sample exam database
exam_schedule = {
    ("CS", "5"): "15 May 2026",
    ("CS", "3"): "10 May 2026",
    ("IT", "5"): "18 May 2026",
    ("IT", "3"): "12 May 2026",
    ("ME", "5"): "20 May 2026"
}

# Entity Extraction Function
def extract_entities(query):

    entities = {
        "semester": None,
        "course": None,
        "date": None
    }

    # Extract Semester
    sem_match = re.search(r'(?:sem|semester)\s*(\d+)', query, re.IGNORECASE)
    if sem_match:
        entities["semester"] = sem_match.group(1)

    # Extract Course / Department
    course_match = re.search(r'\b(CS|IT|ME|CE|EE|EC)\b', query, re.IGNORECASE)
    if course_match:
        entities["course"] = course_match.group(1).upper()

    # Extract Date Words
    date_match = re.search(r'\b(today|tomorrow)\b', query, re.IGNORECASE)
    if date_match:
        entities["date"] = date_match.group(1)

    return entities


# Response Generator
def generate_response(entities):

    semester = entities["semester"]
    course = entities["course"]

    if semester and course:

        key = (course, semester)

        if key in exam_schedule:
            return f"{course} Semester {semester} exam is scheduled on {exam_schedule[key]}."

        else:
            return "Sorry, exam schedule not available for that course and semester."

    elif semester:
        return f"You asked about Semester {semester}. Please specify the course (CS, IT, etc.)."

    elif course:
        return f"You asked about {course}. Please specify the semester."

    else:
        return "Please mention course and semester (example: 'When is SEM 5 CS exam?')."


# Chatbot Loop
def chatbot():

    print("🎓 Exam Query Chatbot")
    print("Type 'exit' to stop.\n")

    while True:

        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Bot: Goodbye! Best of luck for your exams.")
            break

        entities = extract_entities(user_input)

        response = generate_response(entities)

        print("Bot:", response)


# Run chatbot
chatbot()
```
