```python
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
SPELLING_MAP = {
    "pls": "please",
    "plz": "please",
    "u": "you",
    "ur": "your",
    "thx": "thanks",
    "fee's": "fees",
    "timing's": "timings"
}
```

```python
def preprocess_question(text):
    # 1. Lowercasing
    text = text.lower()

    # 2. Spelling normalization
    for wrong, correct in SPELLING_MAP.items():
        text = re.sub(r'\b' + wrong + r'\b', correct, text)

    # 3. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 4. Tokenization
    tokens = word_tokenize(text)

    # 5. Stopword removal
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    return filtered_tokens
```

```python
def preprocess_to_sentence(text):
    tokens = preprocess_question(text)
    return " ".join(tokens)
```

```python
question = "Pls tell me the fee's structure & timings!!!"
processed = preprocess_question(question)
print(processed)
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
from pyspellchecker import SpellChecker

# Downloads
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Objects
lemmatizer = WordNetLemmatizer()
spell = SpellChecker()
stop_words = set(stopwords.words('english'))

SPELLING_MAP = {
    "pls": "please",
    "plz": "please",
    "u": "you",
    "ur": "your",
    "thx": "thanks"
}

def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "V": wordnet.VERB,
        "N": wordnet.NOUN,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

def preprocess_question(text):
    # 1. Lowercase
    text = text.lower()

    # 2. Normalize common student slang
    for wrong, correct in SPELLING_MAP.items():
        text = re.sub(r'\b' + wrong + r'\b', correct, text)

    # 3. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 4. Tokenization
    tokens = word_tokenize(text)

    # 5. Spell correction
    corrected_tokens = []
    for word in tokens:
        if word not in stop_words:
            corrected_tokens.append(spell.correction(word))

    # 6. Lemmatization
    lemmatized_tokens = [
        lemmatizer.lemmatize(word, get_wordnet_pos(word))
        for word in corrected_tokens
        if word not in stop_words
    ]

    return lemmatized_tokens
```
